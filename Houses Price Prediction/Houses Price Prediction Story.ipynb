Dataset: House Prices

House price keeps increasing for a year since pandemic breaks out. I am interested at what features will affect the price of house. Therefore, I have the house price dataset with 79 variables to analyze what variables affect the house price most.  

```python
#Explore the data
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
house = pd.read_csv('train.csv')
```

```python
house.columns
```

```python
#From the above column names, I am intrested at the variales: YearBuilt, GrLivArea, OverallQual, YrSold.
```

```python
sns.scatterplot(x=house['YearBuilt'], y=house['SalePrice'], alpha=0.25)
plt.show()
```

```python
#From the above graph, we can see that the newer the house, the price has a upper trend. 
```

```python
sns.scatterplot(x=house['GrLivArea'], y=house['SalePrice'], alpha=0.25)
plt.show()
```

```python
#From the above graph, we can see that the bigger the house, the price has a upper trend. 
```

```python
sns.scatterplot(x=house['OverallQual'], y=house['SalePrice'], alpha=0.25)
plt.show()
```

```python
#From the above graph, we can see that the nicer the house, the price has a upper trend.
```

```python
sns.scatterplot(x=house['YrSold'], y=house['SalePrice'], alpha=0.25)
plt.show()
```

```python
#From the avoce graph, the house price is relatively stable during 2006 to 2010. 
```

```python
#We will plot a heatmap to see the relationship between saleprice and other variables.
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(house.corr(), cmap='magma')
plt.show()
```

```python
#plot the 20 most important variables instead
plt.figure(figsize=(15,8))
sns.heatmap(house[house.corr().nlargest(20, 'SalePrice')['SalePrice'].index].corr(), cmap='magma', annot=True)
plt.show()
```

```python
#From the above graph, we can see that 'OverallQual' and 'GrLivArea' are the 2 most important variables. 
#We will do more analysis on these 2 variables.
```
